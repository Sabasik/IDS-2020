```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
internet = pd.read_csv("List of Countries by number of Internet Users - Sheet1.csv")
internet
```

```python
users = []
for nr in internet["Internet Users"]:
    users.append(round(int(nr.replace(",", "")), -6))
values, counts = np.unique(users, return_counts=True)
plt.scatter(values, counts, label="Internet Usage")
plt.xlabel("Users")
plt.ylabel("Occurrances")
plt.show()
print(counts)
print(values)
```

```python
pop = []
for nr in internet["Population"]:
    pop.append(round(int(nr.replace(",", "")), -7))
values, counts = np.unique(pop, return_counts=True)
plt.scatter(values, counts, label="Population")
plt.xlabel("People")
plt.ylabel("Occurrances")
plt.show()
print(counts)
print(values)
```
